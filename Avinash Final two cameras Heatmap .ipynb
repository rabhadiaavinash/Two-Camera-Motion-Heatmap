```python
import numpy as np
import cv2
import copy
```

```python
cap = cv2.VideoCapture('part1.mp4')
cap2 =cv2.VideoCapture('part2.mp4')
#cap = cv2.VideoCapture('cctv_part1.mp4')
#cap2 =cv2.VideoCapture('cctv_part2.mp4')
#cap = cv2.VideoCapture('cctv-p1.mp4')
#cap2 =cv2.VideoCapture('cctv-p2.mp4')
#cap = cv2.VideoCapture('short1.mp4')
#cap2 =cv2.VideoCapture('short2.mp4')
#cap = cv2.VideoCapture('cctv_camera_part1.mp4')
#cap2 =cv2.VideoCapture('cctv_camera_part2.mp4')
#cap = cv2.VideoCapture('A.mp4')
#cap2 =cv2.VideoCapture('B.mp4')


#cap3 =cv2.VideoCapture('input.mp4')
#cap4 =cv2.VideoCapture('input.mp4')
```

```python
if (cap.isOpened()== False): 
  print("Error opening video stream or file")
if (cap2.isOpened()== False): 
  print("Error opening video stream or file")
#if (cap3.isOpened()== False): 
#  print("Error opening video stream or file")
#if (cap4.isOpened()== False): 
#  print("Error opening video stream or file")   
```

```python
first_iteration_indicator = 1
background_subtractor = cv2.bgsegm.createBackgroundSubtractorMOG()
i = 0
images = []
```

```python
while(cap.isOpened() & cap2.isOpened() ):
    # Capture frame-by-frame
    ret, frame = cap.read()
    ret, frame1 = cap2.read()
    #ret, frame2 = cap3.read()
    #ret, frame3 = cap4.read()
  
    images.append(frame)
    images.append(frame1)
    #images.append(frame2)
    #images.append(frame3)
  
    if ret == True:
                      
        #stitcher = cv2.Stitcher.create()
        stitcher = cv2.Stitcher_create()
        (status, stitched) = stitcher.stitch(images)
    
        if(status == cv2.STITCHER_OK):
            print("Stitcher OK")
        else:
            print("Stitcher Not OK")
        frame_main = stitched.copy()
        images = []
        
        
        # If first frame
        if first_iteration_indicator == 1:
            
            first_frame = copy.deepcopy(frame_main)
            cv2.imwrite("First_frame_two_camera.jpg",first_frame)
            height, width = frame_main.shape[:2]
            accum_image = np.zeros((height, width), np.uint8)
            #print("accum_image shape:",accum_image.shape)
            first_iteration_indicator = 0
        else:
            
            frame_main = cv2.resize(frame_main, (accum_image.shape[1],accum_image.shape[0]),cv2.INTER_LINEAR)
            
            filter = background_subtractor.apply(frame_main)  # remove the background
                                 
            cv2.imwrite('filter_copy_two_camera.jpg', filter)

            threshold = 2
            maxValue = 2
            ret, th1 = cv2.threshold(filter, threshold, maxValue, cv2.THRESH_BINARY)
                                 
            # add to the accumulated image
            accum_image = cv2.add(accum_image, th1)
           
            cv2.imwrite('mask_copy_two_camera.jpg', accum_image)
         
            # color map applied to mask
            
            color_image_video = cv2.applyColorMap(accum_image, cv2.COLORMAP_HOT)
            
            video_frame = cv2.addWeighted(frame_main, 0.7, color_image_video, 0.7, 0)
      
      
            i+=1
          
                
            cv2.imwrite("Final_frame_two_camera.jpg", video_frame)
             
            cv2.namedWindow("Heatmap video",cv2.WINDOW_NORMAL)  
            cv2.imshow("Heatmap video",video_frame)   
            
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break






cap.release()
cap2.release()
#cap3.release()
#cap4.release()

cv2.destroyAllWindows()     
```
